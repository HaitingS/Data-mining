```python
from sklearn import datasets#引入数据集,sklearn包含众多数据集
from sklearn.model_selection import train_test_split#将数据分为测试集和训练集
from sklearn.neighbors import KNeighborsClassifier
import matplotlib.pyplot as plt
 
###引入数据###
iris=datasets.load_iris()#引入iris鸢尾花数据
iris_X=iris.data#特征变量
iris_y=iris.target#目标值
X_train,X_test,y_train,y_test=train_test_split(iris_X,iris_y,test_size=0.1)#利用train_test_split进行将训练集和测试集进行分开，test_size占10%

knn=KNeighborsClassifier()#引入训练方法
knn.fit(X_train,y_train)#进行填充测试数据进行训练

x=np.arange(0,len(X_test))
yp = knn.predict(X_test)
y = y_test
fig, ax = plt.subplots()
ax.plot(x, y, linewidth=2.0,label="ground-truth",marker="o")
ax.plot(x, yp, linewidth=2.0,label="predict")
ax.legend()
#绘图
# plt.show()

# from sklearn.linear_model import LinearRegression
# linreg = LinearRegression()
# linreg.fit(X_train,y_train)

from sklearn import metrics
# 用scikit-learn计算MSE
print ("accuracy_score:",metrics.accuracy_score(y, yp))

```
